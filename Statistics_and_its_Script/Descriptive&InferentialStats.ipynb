#### Question 5.
In January 2005, a company that monitors Internet traffic (WebSideStory) reported that its sampling revealed that the Mozilla Firefox browser launched in 2004 had grabbed a 4.6% share of the market.

I.	If the sample were based on 2,000 users, could Microsoft conclude that Mozilla has a less than 5% share of the market?

II.	WebSideStory claims that its sample includes all the daily Internet users. If that’s the case, then can Microsoft conclude that Mozilla has a less than 5% share of the market?


**-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------**

**Answer:**

I.

Given: A company which monitors Internet traffic have reported that its sampling have revealed that Mozilla firebox has grabbed 4.6% share of the market<br>
i.e.,Population propotion of share market by Mozilla is P=4.6%=0.046<br>
And also given that sample were based on 2000 users i.e., n=2000<br>

The Hypthesis of interest is:<br>
H0: Morzilla has greater than or equal to 5%/0.05 share of the market (H0:P>=5%)<br>
H1: Morzilla has less than 5%/0.05 share of the market (H1:P<5%)

The test statistic used here will be 'One sample Z-test for propotion'<br>
Z= (P-p)/√((p*(1-p))/n)<br>
where P= Population propotion, p= sample propotion<br>

So,<br>
Z=(0.046-0.05)/√((0.05*(1-0.05))/20000)<br>
Z=-0.821<br>
The p value is 0.2061 (From normal table)<br>

Since, p value is greater than 0.05, we do not reject null hypothesis i.e., Morzilla has greater than or equal to 5%/0.05 share of the market.


```python
# Using python code manually:
from scipy import stats
print('H0: Morzilla has greater than or equal to 5%/0.05 share of the market (H0:P>=5%)')
print('H1: Morzilla has less than 5%/0.05 share of the market (H1:P<5%)')
print('The test statistic is:')
z_score=(0.046-0.05)/(np.sqrt((0.05*(1-0.05))/2000))
print('z-Score=',z_score)
p_value=1-stats.norm.cdf(abs(z_score))
print('The p value is:',p_value)
if p_value>0.05:
    print('Since, p value is greater than 0.05, we do not reject null hypothesis i.e., Morzilla has greater than or equal to 5% share of the market.')
else:
    print('Since, p value is less than 0.05, we reject null hypothesis i.e.,Morzilla has less than 5% share of the market')
```

```python
# Using build in function for propotion test
from statsmodels.stats.proportion import proportions_ztest
statistic,P_value=proportions_ztest(count=92,nobs=2000,value=0.05, alternative='smaller')
print('Z-Statistic:%0.3f, P-value:%0.3f'%(statistic,P_value))
if P_value>0.05:
    print('We do not reject Null hypothesis and conclude that Morzilla has greater than or equal to 5% share of the market')
else:
    print('We reject Null Hypothesis and conclude that Morzilla has less than 5% share of the market ')
```
